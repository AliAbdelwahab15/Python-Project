```python
import pandas as pd
import numpy as np
import seaborn as sns
import statistics as stats
import matplotlib.pyplot as plt
```

```python
class DataProcessor:
    def load_data(self):
        x= pd.read_csv("data/Data.csv")
        return x
    def clean_data(self,read):
        newdata = read.dropna()
        return newdata
data=DataProcessor()
loading=data.load_data()
ndata=data.clean_data(loading)
print(ndata)
```

```python
class DataAnalyzer:
    def get_average_of_column(self,data):
        avr = np.average(data)
        return avr
    def get_distribution_of_column(self,data):
        stdv = np.std(data)
        return stdv
    def get_median_of_column(self,data):
        med = np.median(data)
        return med
    def get_mod_of_column  (self,data):
        mod=stats.mode(data)
        return mod
dataAny = DataAnalyzer()
```

```python
print(dataAny.get_average_of_column(ndata["Goals"]))
```

```python
print(dataAny.get_distribution_of_column(ndata["Matches_Played"]))
```

```python
print(dataAny.get_median_of_column(ndata["Mins"]))
```

```python
print(dataAny.get_mod_of_column(ndata["Goals"]))
```

```python
class DataVisualizer:
    def plot_line_chart(self,x,y):
        line = sns.lineplot(x,y)
        return line
    def plot_distribution(self,x,y,hues):
        bar = sns.barplot(x,y,hue=hues)
        return bar
    def plot_pie(self,x,label):
        p=plt.pie(x,labels=label)
        return p
    def plot_scatter(self,x,y,hues):
        scatter = sns.scatterplot(x,y,hue=hues)
        return scatter
visual= DataVisualizer()
labels= pd.unique(ndata["Country"])
Country=ndata["Country"]
England = len([x for x in Country if x == "England"])
Spain = len([x for x in Country if x == "Spain"])
Italy= len([x for x in Country if x == "Italy"])
Germany = len([x for x in Country if x == "Germany"])
Brazil = len([x for x in Country if x == "Brazil"])
France = len([x for x in Country if x == "France"])
USA = len([x for x in Country if x == "USA"])
Portugal = len([x for x in Country if x == "Portugal "])
Netherlands = len([x for x in Country if x == " Netherlands"])
Country_list= [Spain, Italy, Germany, England, Brazil, France, USA, Portugal, Netherlands]
```

```python
visual.plot_line_chart(ndata["Goals"],ndata["Shots"])
plt.savefig("linechart.png")
```

```python
plt.subplots(figsize=(20,10))
visual.plot_distribution(ndata["Goals"],ndata["Shots"],hues=ndata["Country"])
plt.savefig("distribution.png")
```

```python
plt.subplots(figsize=(10,10))
visual.plot_pie(Country_list,label= labels)
plt.savefig("piechart.png")
```

```python
plt.subplots(figsize=(10,5))
visual.plot_scatter(ndata["Shots"],ndata["xG"],hues=ndata["Country"])
plt.savefig("scatterplot.png")
```
